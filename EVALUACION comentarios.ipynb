```python
# Defining Node class for doubly linked list
class Nodo:
    def __init__(self, Ano_Notificacion, CodLocalidadResidencia, NombreLocalidadResidencia, CicloVital, ClasificaciondelaConducta, SEXO, Casos):
        # Initializes a node with the given data fields and pointers to next and previous nodes.
        self.Ano_Notificacion = Ano_Notificacion
        self.CodLocalidadResidencia = CodLocalidadResidencia
        self.NombreLocalidadResidencia = NombreLocalidadResidencia
        self.CicloVital = CicloVital
        self.ClasificaciondelaConducta = ClasificaciondelaConducta
        self.SEXO = SEXO
        self.Casos = Casos
        self.next = None  # Points to the next node in the list
        self.prev = None  # Points to the previous node in the list

# Defining the doubly linked list class
class ListaDoble:
    def __init__(self):
        # Initializes an empty list with head as None.
        self.head = None

    # Function to insert data into the doubly linked list
    def insertar(self, Ano_Notificacion, CodLocalidadResidencia, NombreLocalidadResidencia, CicloVital, ClasificaciondelaConducta, SEXO, Casos):
        # Creates a new node with the provided data.
        nuevo_nodo = Nodo(Ano_Notificacion, CodLocalidadResidencia, NombreLocalidadResidencia, CicloVital, ClasificaciondelaConducta, SEXO, Casos)
        # If the list is empty, set the new node as the head.
        if not self.head:
            self.head = nuevo_nodo
        else:
            actual = self.head  # Start from the head of the list
            # Traverse to the last node
            while actual.next:
                actual = actual.next
            actual.next = nuevo_nodo  # Set the next pointer of the last node to the new node
            nuevo_nodo.prev = actual  # Set the previous pointer of the new node to the current last node

    # Function to display data from front to back
    def mostrar_adelante(self):
        # Check if the list is empty
        if not self.head:
            print("La lista está vacía.")  # Print message if the list is empty
            return
        actual = self.head  # Start from the head of the list
        contador = 0  # Counter to track how many items are shown
        # Traverse the list and show all data
        while actual:
            print(f"Año: {actual.Ano_Notificacion}, Localidad: {actual.NombreLocalidadResidencia}, Edad: {actual.CicloVital}, Clasificación: {actual.ClasificaciondelaConducta}, Sexo: {actual.SEXO}, Casos: {actual.Casos}")
            actual = actual.next  # Move to the next node
            contador += 1  # Increment counter

    # Function to display data from back to front
    def mostrar_atras(self):
        # Check if the list is empty
        if not self.head:
            print("La lista está vacía.")  # Print message if the list is empty
            return
        actual = self.head  # Start from the head of the list
        # Move to the last node by traversing the list
        while actual.next:
            actual = actual.next
        contador = 0  # Counter to track how many items are shown
        # Traverse the list backwards and show all data
        while actual:
            print(f"Año: {actual.Ano_Notificacion}, Localidad: {actual.NombreLocalidadResidencia}, Edad: {actual.CicloVital}, Clasificación: {actual.ClasificaciondelaConducta}, Sexo: {actual.SEXO}, Casos: {actual.Casos}")
            actual = actual.prev  # Move to the previous node
            contador += 1  # Increment counter

    # Function to filter data and return a new list with the results
    # Filter data only by Notification_Year and ResidenceCod
    def filtrar(self, Ano_Notificacion, CodLocalidadResidencia):
        actual = self.head  # Start from the head of the list
        lista_filtrada = ListaDoble()  # Create a new doubly linked list for filtered data
        encontrado = False  # Flag to check if any matching nodes are found

        # Traverse the list to find nodes that match the given criteria
        while actual:
            # Compare data by Year and Location Code
            if (actual.Ano_Notificacion == Ano_Notificacion and
                actual.CodLocalidadResidencia == CodLocalidadResidencia):
                # Insert matching nodes into the new filtered list
                lista_filtrada.insertar(actual.Ano_Notificacion, actual.CodLocalidadResidencia, actual.NombreLocalidadResidencia,
                                        actual.CicloVital, actual.ClasificaciondelaConducta, actual.SEXO, actual.Casos)
                encontrado = True  # Set flag to True when a match is found
            actual = actual.next  # Move to the next node
        # If no match is found, print a message
        if not encontrado:
            print("No se encontraron coincidencias con los criterios proporcionados.")
        return lista_filtrada  # Return the filtered list

# Function to generate the options for locations, life cycles, classification and sex
def obtener_opciones(df, columna):
    opciones = df[columna].unique().tolist()  # Get unique values from the specified column
    opciones_dict = {i + 1: opciones[i] for i in range(len(opciones))}  # Create a dictionary to map options to numbers
    return opciones_dict

# Function to generate the classification and sex options
def obtener_opciones_fijas(opciones):
    return {i + 1: opciones[i] for i in range(len(opciones))}  # Create a dictionary to map fixed options (classification, sex) to numbers

# Show available options and allow the user to choose
def seleccionar_opcion(opciones, nombre_columna):
    print(f"\nSeleccione un {nombre_columna}:")  # Prompt the user to select an option
    print('''
    1. Usaquén
    2. Chapinero
    3. Santa Fe
    4. San Cristóbal
    5. Usme
    6. Tunjuelito
    7. Bosa
    8. Kennedy
    9. Fontibón
    10. Engativá
    11. Suba
    12. Barrios Unidos
    13. Teusaquillo
    14. Los Mártires
    15. Antonio Nariño
    16. Puente Aranda
    17. La Candelaria
    18. Rafael Uribe
    19. Ciudad Bolívar
    20. Sumapaz
     ''')
    seleccion = int(input(f"Ingrese el número correspondiente a la {nombre_columna.lower()}: "))
    return opciones.get(seleccion)  # Return the selected option

# Loading data from the DataFrame
import pandas as pd  # Import pandas to handle CSV file reading

# Path to CSV file
file_path = 'DATOS.csv'  # Specify the path to the CSV file

# Read CSV file
df = pd.read_csv(file_path, encoding='latin1', delimiter=';')  # Read the CSV file using pandas

# Create an instance of the doubly linked list
lista_doble = ListaDoble()  # Create a new doubly linked list

# Insert data from DataFrame into doubly linked list
for _, fila in df.iterrows():
    lista_doble.insertar(fila['Ano_Notificacion'], fila['CodLocalidadResidencia'], fila['NombreLocalidadResidencia'], fila['CicloVital'], fila['ClasificaciondelaConducta'], fila['SEXO'], fila['Casos'])  # Insert each row of the CSV into the list

# User Options Menu
def menu():
    localidades_opciones = obtener_opciones(df, 'CodLocalidadResidencia')  # Generate location code options from the data

    while True:
        print("\nMenu:")  # Display menu options
        print("1. Mostrar datos filtrados (de adelante hacia atrás y de atrás hacia adelante)")
        print("2. Salir")

        opcion = input("Seleccione una opción (1-2): ")  # Take user input for menu selection

        if opcion == '1':
            ano = int(input("Ingrese el año de notificación (ej. 2012): "))  # Prompt the user for the year
            localidad = seleccionar_opcion(localidades_opciones, 'Código de Localidad')  # Let the user select a location code

            # Filter data and display front to back and back to front
            lista_filtrada = lista_doble.filtrar(ano, localidad)  # Filter the data based on user inputs
            print("\nDatos filtrados de adelante hacia atrás:")
            lista_filtrada.mostrar_adelante()  # Display filtered data from front to back
            print("\nDatos filtrados de atrás hacia adelante:")
            lista_filtrada.mostrar_atras()  # Display filtered data from back to front
            print("\n")
            print("Imprimiendo 10 datos de la tabla")
            print("\n")
            print(df.head(10))

        elif opcion == '2':
            print("Saliendo del programa.")  # Exit the program
            break

        else:
            print("Opción no válida. Intente nuevamente.")  # Handle invalid option input

# Run the menu
menu()  # Execute the menu function to start the program "
```
